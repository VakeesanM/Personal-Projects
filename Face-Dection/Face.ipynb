```python
import pandas as pd
import numpy as np
import json
```

```python
file_path = r"D:\User\Vakeesan's Pc\Download\archive\face_detection.json"
```

```python
data = []
with open(file_path, 'r', encoding='utf-8') as f:
    for line in f:
        data.append(json.loads(line.strip()))

```

```python
rows = []
for item in data:
    image_url = item['content']
    annotations = item['annotation']
    for ann in annotations:
        label = ann['label']
        notes = ann['notes']
        points = ann['points']
        imageWidth = ann['imageWidth']
        imageHeight = ann['imageHeight']
        
        rows.append({
            'content': image_url,
            'label': label,
            'notes': notes,
            'x1': points[0]['x'],
            'y1': points[0]['y'],
            'x2': points[1]['x'],
            'y2': points[1]['y'],
            'imageWidth': imageWidth,
            'imageHeight': imageHeight
        })

```

```python
df = pd.DataFrame(rows)

print(df.head())
```
